# Regression plot 

```python
#The normal imports
import numpy as np
import pandas as pd
from numpy.random import randn

#Import the statslibrary from numpy
from scipy import stats 

#These are the plotting modules and libraries we will use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Command so that plots appear in the ipython Notebook
%matplotlib inline
```

```python
# loading datasets from seaborn 

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot('total_bill', 'tip', tips)
# regressionplot(x,y,data)
```

```python
sns.lmplot('total_bill', 'tip', tips,
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# change the order and not keep linear 
# for higher orders add order=4 or any number
sns.lmplot('total_bill','tip', tips, order = 3,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips, fit_reg=False)
```

```python
# making a tip percentage column
tips['tip_perc'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
#compare size and perc

sns.lmplot('size', 'tip_perc',tips)
```

```python
# to get more clear picture use jitter to spread the data more 
sns.lmplot('size', 'tip_perc',tips, x_jitter=0.1 )
```

```python
#estimate the variance in each bins
sns.lmplot('size', 'tip_perc', tips, x_estimator= np.mean)
```

```python
# introducing multiple regression plots based on gender one male and other for female by using hue argument 
sns.lmplot('total_bill', 'tip_perc',tips, hue='sex', markers=['o','x'])
```

```python
sns.lmplot('total_bill', 'tip_perc',tips, hue='day')
```

```python
# fit a regression line
sns.lmplot('total_bill', 'tip_perc', tips, lowess=True, line_kws={'color':'blue'})
```

```python
sns.regplot('total_bill','tip_perc',tips)
```

```python
# lets create two plots 

fig, (axis1, axis2) = plt.subplots(1,2,sharey = True)

sns.regplot('total_bill', 'tip_perc',tips, ax=axis1)
sns.violinplot(tips['tip_perc'], tips['size'], color='Red', ax=axis2)
```

# heatmap and cluster matrices

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot('month','year','passengers')
flight_dframe
# pivot_function(rows, columns, data to be filled)
# you can't redo the pivot function again(REMEMBER)
```

```python
# this data will help us to make heatmap
sns.heatmap(flight_dframe)
```

```python
# filling the values in each grid use annot and to properly size the font use fmt='d')
sns.heatmap(flight_dframe, annot=True, fmt='d')
```

```python
# change color and stlye
sns.heatmap(flight_dframe, center=flight_dframe.loc['January', 1955])
# instead of getting a single color place center as some year value and then change color, 
#white being the mid value of number of passengers in 1955
```

```python
fig, (axis1, axis2) = plt.subplots(2,1)
# subplots(row, column), here it means two graphs one below the other

yearly_flights =flight_dframe.sum()
# here we will get years as our index and sum of flights as our value 

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

# taking only the index of yearly_flights for exmaple 1995

flights= pd.Series(yearly_flights.values)
flights =pd.DataFrame(flights)
# taking the actual sum values of yearly_flights for exmaple 600

year_dframe = pd.concat((years, flights), axis=1)
year_dframe.columns = ['Years', 'Flights']

sns.barplot('Years', y ='Flights', data=year_dframe, ax=axis1)

sns.heatmap(flight_dframe, cmap = 'Blues', ax=axis2, cbar_kws={'orientation':'horizontal'})
```

```python
# cluster map

sns.clustermap(flight_dframe)
```

```python
sns.clustermap(flight_dframe, col_cluster=False)
```

```python
sns.clustermap(flight_dframe, standard_scale=1)
# standardize by columns
```

```python
sns.clustermap(flight_dframe, standard_scale=0)
# standardize by rows
```

```python
sns.clustermap(flight_dframe, z_score=1)
```
